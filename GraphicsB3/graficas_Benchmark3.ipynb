## TASK 2: Benchmark of matrix multiplication

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Matrix Type': ['Sparce', 'Sparce', 'Sparce', 'Sparce', 'Sparce', 'Dense', 'Dense', 'Dense', 'Dense', 'Dense'],
    'Tama침o de la Matriz': [130, 420, 662, 1074, 1473, 130, 420, 662, 1074, 1473],
    'Tiempo de Ejecuci칩n': [4, 13, 14, 53, 151, 
                            8, 263, 1426, 8834, 33292]
}

df = pd.DataFrame(data)
print(df)
df.to_csv('datos.csv', index=False)
```

```python
# Bar chart to compare execution times between languages
plt.figure(figsize=(10, 6))
plt.bar(df['Matrix Type'], df['Tiempo de Ejecuci칩n'], color=['blue', 'green', 'red'])
plt.xlabel('Matrix Type')
plt.ylabel('Execution time (seconds)')
plt.title('Execution Time Comparison by Matrix Multiplication Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.savefig('bar_chart.png')
plt.show()

# Line graph to show trend of execution times with respect to the size of the matrix
plt.figure(figsize=(10, 6))
for lang in df['Matrix Type'].unique():
    df_lang = df[df['Matrix Type'] == lang]
    plt.plot(df_lang['Tama침o de la Matriz'], df_lang['Tiempo de Ejecuci칩n'], label=lang)
plt.xlabel('Matrix Size')
plt.ylabel('Execution time (seconds)')
plt.title('Run Time Trend by Matrix Size')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('line_chart.png')
plt.show()
```

```python
matrix_names = ['20%', '40%', '60%', '80%']
dense_runtimes = [193, 179, 164, 156]
sparse_runtimes = [68, 57, 20, 18]

plt.figure(figsize=(10, 6))
plt.plot(matrix_names, sparse_runtimes, marker='o', label='Sparse Matrix')
plt.plot(matrix_names, dense_runtimes, marker='o', label='Dense Matrix')
plt.xlabel('Percentage of Zeros')
plt.ylabel('Execution Time (milliseconds)')
plt.title('Matrix Multiplication Execution Time per Matrix Percentage of Zeros')
plt.grid(True)
plt.legend()
plt.savefig('line_chart2.png')
plt.show()
```
